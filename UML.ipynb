```python
import pandas as pd
from pathlib import Path
```

```python
# Load the Data
file_path = Path("./Data/myopia.csv")
myo = pd.read_csv(file_path)
myo.head(10)
```

```python
# Check if any values are non nemueric values
myo.dtypes
```

```python
# Remove the 'Myopic' column
myo_fixed = myo.drop(columns=["MYOPIC"])
myo_fixed.head(30)
```

```python
#     print(f"Duplicate entries: {myo_fixed.duplicated().sum()}")
#     With no duplicates, there is no need to remove them or a need to transform non digit values
for column in myo_fixed.columns:
    print(f"Column {column} has {myo_fixed[column].isnull().sum()} null values")
```

```python
# Scale the data closer together
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_data = scaler.fit_transform(myo_fixed[['AGE','SPHEQ','AL','ACD','LT','VCD','SPORTHR','READHR','COMPHR','STUDYHR','TVHR','DIOPTERHR','MOMMY','DADMY']])
scaled_data
```

```python
# Check scaled data
new_myo = pd.DataFrame(scaled_data, columns=myo_fixed.columns)
new_myo.head()
```

```python
# 
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.90)
myo_pca = pca.fit_transform(new_myo)
pca.explained_variance_ratio_
```

```python
# Insert the data into a dataframe and change the names in something 
myo_pca_df = pd.DataFrame(data=myo_pca, columns=["PCA01", "PCA02", "PCA03", "PCA04","PCA05", "PCA06", "PCA07", "PCA08","PCA09", "PCA10"])
myo_pca_df.head()
```

```python
# Use TSNE to transform the 'Myopic' column we dropped earlier
from sklearn.manifold import TSNE
labels = myo['MYOPIC']
tsne = TSNE(learning_rate = 75)
tsne_features = tsne.fit_transform(myo_pca)
tsne_features.shape
```

```python
# Plot X and Y axis on a plot chart using the transformed TSNE data
import matplotlib.pyplot as plt
myo_pca_df['x'] = tsne_features[:,0]
myo_pca_df['y'] = tsne_features[:,1]
plt.scatter(myo_pca_df['x'], myo_pca_df['y'])
plt.show
```

```python
# Model the data using 10 clusters for the 10 columns, fit the data, then add the predicted model to the dataframe as the 'class' column
from sklearn.cluster import KMeans

model = KMeans(n_clusters=10, random_state=42)
model.fit(myo_pca_df)

predictions = model.predict(myo_pca_df)

myo_pca_df["class"] = model.labels_
myo_pca_df.head()
```

```python
# Use a for loop to make a dataframe about the inertia to find out where the elbow curve is to find the best amount of clusters
inertia = []
k = list(range(1, 11))

for i in k:
    km = KMeans(n_clusters = i, random_state = 42)
    km.fit(myo_pca_df)
    inertia.append(km.inertia_)
    
elbow_curve_data = {"k": k, "inertia": inertia}
elbow_df = pd.DataFrame(elbow_curve_data)
elbow_df.head()
```

```python
# Plot the for looped data to visualize the elbow curve
plt.plot(elbow_df['k'], elbow_df['inertia'])
plt.xticks(range(1,11))
plt.xlabel("Cluster Numbers")
plt.ylabel("Inertia")
plt.title("Elbow Curve")
plt.show()
```

```python
# The point the elbow seems to plateau around 5 so that'll be the amount of clusters used to model the cluster data
model = KMeans(n_clusters=5, random_state=42)
model.fit(myo_pca_df)
predictions = model.predict(myo_pca_df)
```

```python
# Model the new 5 clustered data to see if the data can be clustered comfortably into 5 groups
def get_clusters(k, data):
    model = KMeans(n_clusters=k, random_state=42)
    model.fit(myo_pca_df)
    predictions = model.predict(myo_pca_df)
    myo_pca_df["class"] = model.labels_
    return myo_pca_df
```

```python
# The previusly modeled 10 clusters from the class column changed to fit the 5 clusters from the elbow curve peak
clusters = get_clusters(5, myo_pca_df)
clusters
```

```python
# Plot the 5 clustered data to visualize the different groups
def show_clusters(df):
    plt.scatter(myo_pca_df['x'], myo_pca_df['y'], c=myo_pca_df["class"])
    plt.show()

show_clusters(clusters)
```
